```python
#11. Kiểm tra có giá trị NaN nào trong DataFrame không

nan_columns = df.isnull().sum()
print(nan_columns)
```

```python
#12. # Loại bỏ các hàng trùng lặp dựa trên cột 'Name', giữ lại hàng đầu tiên
df_unique = df.drop_duplicates(subset='Name', keep='first')
print(df_unique)
```

```python
#  13. Thay thế các giá trị của cột <Age> lớn hơn 40 bằng giá trị trung bình của cột đó

average_age = df['Age'].mean()

df['Age'] = df['Age'].apply(lambda x: average_age if x > 40 else x)
print(df)
```

```python
#14.Tính chỉ số BMI và thêm vào cột <cs_BMI>
df['cs_BMI'] = df['Weight'] / (df['Height'] ** 2)
print(df)

```

```python
# 15. Theo thang đo kết quả tính BMI của châu Á: Chỉ số BMI dưới 18,5 là thiếu cân, 18.5-22,9 là bình thường, 23-24,9 là thừa cân, từ 25 trở lên là béo phì. Tạo cột với tên <Phan_Loai> với các mức ‘0’,’1’,’2’,’3’ tương ứng với mức chỉ số BMI trên.
def classify_bmi(bmi):
    if bmi < 18.5:
        return 0  # Thiếu cân
    elif 18.5 <= bmi <= 22.9:
        return 1  # Bình thường
    elif 23 <= bmi <= 24.9:
        return 2  # Thừa cân
    else:
        return 3  # Béo phì

# Thêm cột <Phan_Loai> vào DataFrame
df['Phan_Loai'] = df['cs_BMI'].apply(classify_bmi)
print(df)
```

```python
#16. Chuẩn hóa giá trị của cột <cs_BMI> chuẩn hóa giá trị trong đoạn từ 0 tới 1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['cs_BMI_normalized'] = scaler.fit_transform(df[['cs_BMI']])
print(df)
```

```python
#17.Cho biết: + ‘Mr‘tương ứng với ‘male’, + ’Mrs’ và ‘Miss’ tương ứng với ‘female’. Tạo cột với tên <GT>, Sau khi thực hiện, xóa các từ Mr, Mrs, Miss nếu trong cột <Name> còn các từ đó.

df['GT'] = df['Name'].apply(lambda x: 'male' if 'Mr.' in x else 'female')
df['Name'] = df['Name'].str.replace('Mr. ', '').str.replace('Mrs. ', '').str.replace('Miss. ', '')
print(df)
```

```python
#18.Vẽ biểu đồ tròn cho <GT>.
import matplotlib.pyplot as plt
gender_counts = df['GT'].value_counts()
gender_labels = gender_counts.index
gender_sizes = gender_counts.values

plt.figure(figsize=(8, 8))
plt.pie(gender_sizes, labels=gender_labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Pie Chart of Gender Distribution')
plt.show()
```

```python
#19. Trực quan hóa dữ liệu người nữ trên <Phan_Loai>.

def classify_bmi(bmi):
    if bmi < 18.5:
        return 0  
    elif 18.5 <= bmi <= 22.9:
        return 1  # Bình thường
    elif 23 <= bmi <= 24.9:
        return 2  # Thừa cân
    else:
        return 3  # Béo phì

df['Phan_Loai'] = df['cs_BMI'].apply(classify_bmi)

# Lọc dữ liệu người nữ
female_data = df[df['GT'] == 'female']

# Vẽ biểu đồ phân bố dữ liệu người nữ trên <Phan_Loai>
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(female_data['Name'], female_data['Phan_Loai'], color='purple')
plt.xlabel('Name')
plt.ylabel('Phan_Loai')
plt.title('Phan Loai của Người Nữ Dựa Trên Chỉ Số BMI')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
